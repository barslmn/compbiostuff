## Linux Kabuk (Shell)
Kabuk, komutlar ile işletim sistemi ile etkileşmemizi sağlayan bir kullanıcı arayüzüdür.
Bu arayüzü sağlayan farklı yazımlar vardır bash, zsh, dash v.b.

Bu kabuk yazılımlarınaa, grafik arayüz sağlayan uçbirimi(terminal) yazılımları ile ya da ttyler üzerinden ulaşabiliriz. Örnek uçbirim yazılımları: urxvt, st, termite.

Komut satırında çalışan yazılımlar linux standart girdi ve standart çıktısı ile etkileşebilir.


### Veri indirme

**wget komutu**

wget komutu kullanarak dosya indirme.

````python
%%bash
wget -q https://www.gutenberg.org/files/2600/2600-0.txt -O WarAndPeace
````

### Yeni satır karakterlerini düzelt
**tr komutu** (**tr**anslate)

Dos dosya tipinde bulunan '\r' karakterlerini sil.

**> operatörü**

**mv komutu** (**m**o**v**e)

Çıkan dosyayı geri yerine taşı.

````python
%%bash
tr -d '\r' < WarAndPeace > tmp
mv tmp WarAndPeace
````

### Verinin Önhazırlığı

**sed komutu** (**s**tream **ed**itor)

Sed komutu dosya içerisindeki ya da standart girdi çıktı da farklı değişiklikler yapmaya olanak sağlar.

#### Epilog sonrası satırları kullanma
```
sed '/^FIRST*/Q' WarAndPeace
```

#### Chapter başlıklarını sil
```
sed '/^CHAPTER/d' WarAndPeace
```

#### Boş satırları sil
```
sed '/^$/d' WarAndPeace
```

#### Alfanumerik olmayan karakterleri değiştir
```
sed 's/[^a-zA-Z0-9]//g' WarAndPeace
```

#### Komutların birbirine bağlanarak kullanılımı (pipe)
```
sed '/^FIRST*/Q' WarAndPeace | sed '/^CHAPTER/d' | sed '/^$/d' | sed 's/[^a-zA-Z0-9]//g'
```


````python
%%bash
sed '/^FIRST*/Q' WarAndPeace | sed '/^CHAPTER/d' | sed '/^$/d' | sed 's/[^a-zA-Z0-9]//g' > tmp
mv tmp WarAndPeace
````

#### Birinci kitap öncesi satırları kullanma

**cat komutu** (con**cat**enate)

Birden fazla dosyayı ardarda eklemeye yarar.

**tac komutu** (cat komutunun tersi)

cat komutu gibi ancak dosyayı sonundan başına doğru okur.

```
tac WarAndPeace | sed '/^BOOKONE/q' | tac 
```

**&& operatörü**
Operatörü ile birden fazla komut aynı satırda yazılabilir. İlk komutun çalışması bittince sonraki komut çalışmaya başlar.

````python
%%bash
tac WarAndPeace | sed '/^BOOKONE/q' | tac > tmp && mv tmp WarAndPeace
````

### Python3
Python3, **genel amaçlı**(arayüz geliştirme, web sunucusu, hesaplama, betikleme -scripting- gibi birçok farklı işte kullanılması),

**yüksek seviye** (daha kolay anlaşılabilir olması),

**dinamik yazılan**-dynamicly typed- (değişkenlerin ne olduklarının belirtilmesine gerek olmayan),

**yorumlanan**-interpreted-(çalıştırıldığında makine koduna dönüştürülen) bir yazılım dilidir.

````python
# Dosyayı python3 ile açmak.
with open('WarAndPeace') as f:
    lines = f.read().splitlines()

# Shelldeki grep komutunun replikası
for line in lines:
    if line.startswith('BOOK'):
        print(line)    
````

````python
# Boş listeyi kitap başlıkları ile doldurma
book_titles = []
for line in lines:
    if line.startswith('BOOK'):
        book_titles.append(line)

# Boş bir sözlüğe, başlıkları anahtar olarak ekleme
books = {}
for title in book_titles:
    books[title] = ''

# Sözlüğe başlıkların altındaki metni değer olarak ekleme
key = 'key'
for line in lines:
    if line.startswith('BOOK'):
        key = line
        continue
    books[key] += line
````

````python
# Kitaplar içerisindeki karakter sayımı
for title, book in books.items():
    print(title, '\t', len(book))
````

````python
import random
import string


def mutated(mutation_rate):
    r = random.randint(1,100)
    if r <= mutation_rate:
        return True

    
def mutate_books(books, mutation_rate):
    mutated_books = {}
    for title, book in books.items():
        mutated_book = ''
        for character in book:
            if mutated(mutation_rate):
                possible_mutations = string.ascii_uppercase + string.ascii_lowercase + string.digits
                possible_mutations = possible_mutations.replace(character, '')
                mutant = random.choice(possible_mutations)
                mutated_book += mutant
            else:
                mutated_book += character
        mutated_books[title] = mutated_book
    return mutated_books
````

````python
m_books = mutate_books(books, 10)
for wt, mt in zip(books['BOOKONE1805'][:100], m_books['BOOKONE1805'][:100]):
    if wt == mt:
        print(wt, '\t-', '\t', mt)
    else:
        print(wt, '\tX', '\t', mt)
````
